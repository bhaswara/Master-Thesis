```python
import glob
import os.path
import random
import numpy as np
import os
import json
```

```python
#get the data address of the images
k = 0
INPUT_DATA = r'/home/irfandw/Works/dataset/data_aligned_resize/FRGC/Testing/Subjects/'
sub_dirs = [x[0] for x in os.walk(INPUT_DATA)]
sub_dirs[1:]
```

```python
#Create genuine pairs from the subjects 
#pair_two_list = [[0, 1],[0,2],[0,3],[1,2],[1,3],[2,3]]
b = 0
w = []
for subdir in sub_dirs[1:]:
    extensions = 'png'
    file_list = []
    dir_name = os.path.basename(subdir)
    #print(dirname)
    file_glob = os.path.join(INPUT_DATA, dir_name, '*.' + extensions)
    #print(file_glob)
    file_list.extend(glob.glob(file_glob))
    if not file_list: continue
    
    #print(file_list)
    pair_two_list = [[0, 1],[0,2],[0,3],[1,2],[1,3],[2,3]]
    label_name = dir_name
    length = len(file_list)
    with open("pair_FRGC_test_genuine.txt", "a") as f:
        for k in pair_two_list:
            base_name1 = os.path.basename(file_list[k[0]])
            base_name2 = os.path.basename(file_list[k[1]])
            print(label_name + "\t" + base_name1 + "\t" + base_name2 + "\n")
            f.write(label_name + "\t" + base_name1 + "\t" + base_name2 + "\n")
            w.append(base_name1)
            b = b+1

print(b)
```

```python
#Shuffle genuine pairs

with open('pair_FRGC_test_genuine.txt','r') as source:
    data = [ (random.random(), line) for line in source ]
    data.sort()
    
with open('shufflepairs_FRGC_test_genuine.txt','w') as target:
    for _, line in data:
        target.write( line )
```
